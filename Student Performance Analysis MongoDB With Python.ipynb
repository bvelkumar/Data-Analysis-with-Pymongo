```python
#Import the Pymongo

from pymongo import MongoClient
```

```python
#Connect With MongoDB 

client = MongoClient('mongodb://localhost:27017/')
```

```python
#List of Database Names 

client.list_database_names()
```

```python
#Db details 

db = client ["studentperformance"]
```

```python
#Collection detail

db.list_collection_names()
```

```python
#To Count the total no of documents 

db.studentdata.count_documents({ })
```

```python
#To Find the one document 

db.studentdata.find_one()
```

```python
#Getting All Documents 

for studentperformance in db.studentdata.find({}):
    print(studentperformance)
```

```python
#Male wise Average Score Performance in overall percentage


b=db.studentdata.aggregate( [
   { "$match": { "gender": "male" } },
   { "$group": { "_id": "$male", "Malewisescoreperformance": { "$avg": "$overall score in percentage" } } }
] )

for i in b:
    print(i)  
```

```python
#Female wise Average Score Performance in overall percentage


b=db.studentdata.aggregate( [
   { "$match": { "gender": "female" } },
   { "$group": { "_id": "$female", "Femalewisescoreperformance": { "$avg": "$overall score in percentage" } } }
] )

for i in b:
    print(i) 
```

```python
#Group Wise Average Score Performance race/ethnicity()

r=["group A","group B","group C","group D","group E"]
for i in r:
    b=db.studentdata.aggregate([{"$match":{"race/ethnicity":i}},{"$group":{"_id":i,"avg":{"$avg":"$overall score in percentage"}}}])
    for j in b:
        print(j)
```

```python
#parental level of education Average Score Performance


ple=["associate's degree","bachelor's degree","high school","master's degree","some college","some high school"]
for i in ple:
    b=db.studentdata.aggregate([{"$match":{"parental level of education":i}},{"$group":{"_id":i,"avg":{"$avg":"$overall score in percentage"}}}])
    for j in b:
        print(j)
```

```python
#lunch wise Average Score Performance

luw=["free/reduced","standard"]
for i in luw:
    b=db.studentdata.aggregate([{"$match":{"lunch":i}},{"$group":{"_id":i,"avg":{"$avg":"$overall score in percentage"}}}])
    for j in b:
        print(j)

```
